```python
# encoding the categirical variables
```

```python
# Replace 'Male' with 1 and 'Female' with 0
df1['sex'] = df1['sex'].map({'Male': 1, 'Female': 0})

# List of other binary columns to encode (Yes=1, No=0)
binary_columns = [
    'tb_prior', 'tb_prior_Pul','hemoptysis', 'weight_loss', 'smoke_lweek', 'fever', 'night_sweats'
]

# Apply mapping to these columns
for col in binary_columns:
    df1[col] = df1[col].map({'Yes': 1, 'No': 0})
```
