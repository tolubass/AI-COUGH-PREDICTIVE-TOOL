```python
import os
import pandas as pd

# === Step 1: Load merged metadata CSV ===
csv_path = r"C:\Users\hp\Desktop\THE_AI_POWERED_COUGH_ANALYZER\labelled_tb_meged\df_merged_tb_label.csv"
df_meta = pd.read_csv(csv_path)

# Clean 'filename' column and extract normalized recording name
def normalize_name(name):
    parts = name.split('-', 1)  # remove timestamp prefix
    return parts[1].replace('.wav', '') if len(parts) > 1 else name.replace('.wav', '')

df_meta['normalized_recording'] = df_meta['filename'].astype(str).apply(normalize_name)

# === Step 2: Load spectrogram PNGs ===
spectrogram_dir = r"C:\Users\hp\Desktop\THE_AI_POWERED_COUGH_ANALYZER\best_used\mel_spectrograms_best"
all_png_files = [f.strip() for f in os.listdir(spectrogram_dir) if f.endswith('.png')]

# Normalize PNG filenames
def normalize_png(name):
    parts = name.split('-', 1)
    return parts[1].replace('.png', '') if len(parts) > 1 else name.replace('.png', '')

# === Step 3: Match PNGs to metadata using normalized names ===
matched_rows = []
for png_name in all_png_files:
    png_norm = normalize_png(png_name).split('_')[0]  # only 'recording-1'
    for _, row in df_meta.iterrows():
        if row['normalized_recording'] == png_norm:
            matched_rows.append({
                'participant': row['participant'],
                'filename': row['filename'],
                'normalized_recording': row['normalized_recording'],
                'image_filename': png_name,
                'image_path': os.path.join(spectrogram_dir, png_name),
                'sound_prediction_score': row['sound_prediction_score'],
                'tb_status': row['tb_status']
            })

# === Step 4: Save final matched DataFrame ===
df_final = pd.DataFrame(matched_rows)

print("\n✅ Value counts for TB status:")
print(df_final['tb_status'].value_counts())

print("✅ Total labeled spectrograms matched:", len(df_final))

# Save to CSV
output_csv = r"C:\Users\hp\Desktop\THE_AI_POWERED_COUGH_ANALYZER\cnn_ready_metadata.csv"
df_final.to_csv(output_csv, index=False)
print(f"📁 Saved final labeled spectrogram metadata to: {output_csv}")
```

```python
# === Load your merged metadata CSV ===
csv_path = r"C:\Users\hp\Desktop\THE_AI_POWERED_COUGH_ANALYZER\labelled_tb_meged\df_merged_tb_label.csv" 
df_meta = pd.read_csv(csv_path)

# === Clean 'filename' column to remove .wav extension ===
df_meta['base_filename'] = df_meta['filename'].apply(lambda x: x.replace('.wav', ''))

# === Load all spectrogram PNGs ===
spectrogram_dir = r"C:\Users\hp\Desktop\THE_AI_POWERED_COUGH_ANALYZER\best_used\mel_spectrograms_best"
all_png_files = os.listdir(spectrogram_dir)
all_png_files = [f for f in all_png_files if f.endswith('.png')]

# === Build new DataFrame to hold matched images and labels ===
matched_rows = []

for png_name in all_png_files:
    for _, row in df_meta.iterrows():
        if png_name.startswith(row['base_filename']):
            matched_rows.append({
                'participant': row['participant'],
                'filename': row['filename'],
                'image_filename': png_name,
                'image_path': os.path.join(spectrogram_dir, png_name),
                'sound_prediction_score': row['sound_prediction_score'],
                'tb_status': row['tb_status']
            })

df_final = pd.DataFrame(matched_rows)
```

```python
# === Final checks ===
print(df_final['tb_status'].value_counts())
print("Total labeled spectrograms matched:", len(df_final))

```

# merging wav2vec 2 embedding data with tb status

```python
import pandas as pd

# === Load Files ===
wav2vec_path = r"C:\Users\hp\Desktop\THE_AI_POWERED_COUGH_ANALYZER\wav2vec2_embeddings.csv"
cnn_meta_path = r"C:\Users\hp\Desktop\THE_AI_POWERED_COUGH_ANALYZER\cnn_ready_metadata.csv"

df_wav = pd.read_csv(wav2vec_path)
df_meta = pd.read_csv(cnn_meta_path)

# === Step 1: Clean the filename in wav2vec ===
# Extract base_filename from full filename
df_wav['base_filename'] = df_wav['filename'].apply(
    lambda x: x.split('_modclean')[0].replace('.wav', '').strip()
)

# === Step 2: Merge using base_filename ===
merged_df = pd.merge(df_wav, df_meta[['base_filename', 'tb_status']], on='base_filename', how='inner')

# === Step 3: Save the merged DataFrame ===
save_path = r"C:\Users\hp\Desktop\THE_AI_POWERED_COUGH_ANALYZER\wav2vec2_with_labels.csv"
merged_df.to_csv(save_path, index=False)

print(" Merged shape:", merged_df.shape)
print(" Saved to:", save_path)
print(" Sample rows:")
print(merged_df.head())
```
