```python
import pandas as pd

# Load both CSVs
audio_feat_path = r"C:\Users\hp\Desktop\solicited_data\solicited_wav2vec2_embeddings.csv"


clinical_meta_path = r"C:\Users\hp\Desktop\solicited_data\CODA_TB_Solicited_Meta_Info.csv"



# Read the data
df_audio = pd.read_csv(audio_feat_path)
df_clinical = pd.read_csv(clinical_meta_path)

# Strip or clean filenames if necessary (e.g., remove extra whitespaces)
df_audio['filename'] = df_audio['filename'].astype(str).str.strip()
df_clinical['filename'] = df_clinical['filename'].astype(str).str.strip()

# Merge based on filename
df_merged = pd.merge(df_audio, df_clinical, on='filename', how='inner')  # only keep matched rows

# Save the merged DataFrame
output_path = r"C:\Users\hp\Desktop\solicited_data\merged_matched_solicited_audio_clinical.csv"
df_merged.to_csv(output_path, index=False)

print(f"✅ Successfully merged {df_merged.shape[0]} rows and saved to:\n{output_path}")
```

```python
import pandas as pd

# File paths
clinical_path = r"C:\Users\hp\Desktop\THE_AI_POWERED_COUGH_ANALYZER\CODA_TB\CODA_TB_Clinical_Meta_Info.csv"
audio_path = r"C:\Users\hp\Desktop\solicited_data\merged_matched_solicited_audio_clinical.csv"
output_path = r"C:\Users\hp\Desktop\solicited_data\merged_audio_clinical_matched_only.csv"

# Load the datasets
df_clinical = pd.read_csv(clinical_path)
df_audio = pd.read_csv(audio_path)

# Merge on 'participant' using inner join (only keep matching rows)
merged_df = pd.merge(df_audio, df_clinical, on='participant', how='inner')

# Display number of matched participants
print(f"✅ Number of matched participants: {merged_df.shape[0]}")

# Display list of matched participant IDs (optional)
print("\nMatched Participant IDs:")
print(merged_df['participant'].unique())

# Save merged DataFrame
merged_df.to_csv(output_path, index=False)
print(f"\n✅ Merged file saved to: {output_path}")
```
